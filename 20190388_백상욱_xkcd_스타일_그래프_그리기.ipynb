```python
import matplotlib.pyplot as plt
import numpy as np
with plt.xkcd():
    plt.title("XKCD style plot!!")
    X = np.linspace(-np.pi,np.pi,256)
    C = np.cos(X)
    plt.plot(X,C)
    plt.scatter([np.pi*2/3],[np.cos(np.pi*2/3)],color='r',lw=3)
    plt.annotate('-0.5 Here!',xy=(np.pi*2/3,np.cos(np.pi*2/3)),xycoords='data',xytext=(0.5,0.05),textcoords="axes fraction",fontsize=16,arrowprops=dict(arrowstyle="->",linewidth=1))
```

```python
y = []
for i in range(15):
    if i < 10:
        y.append(5)
    else:
        y.append(15-i)
x = np.arange(0,15)
with plt.xkcd():
    plt.text(0,-1,"\"Stove Ownership\" from xkcd by Randall Monroe")
    plt.plot(x,y)
    plt.ylim(0,7)
    plt.xlabel('time')
    plt.ylabel('my overall health')    
    plt.gca().axes.xaxis.set_ticks([])
    plt.gca().axes.yaxis.set_ticks([])
    plt.gca().axes.spines['right'].set_visible(False)
    plt.gca().axes.spines['top'].set_visible(False)
    plt.annotate('THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',xy=(10,5),xycoords='data',xytext=(0.15,0.43),textcoords="axes fraction",fontsize=15,arrowprops=dict(arrowstyle="->",linewidth=1))
    plt.show()
   
```
