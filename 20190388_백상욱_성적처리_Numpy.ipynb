```python
import numpy as np
import matplotlib.pyplot as plt

score = np.array([[75,90,55,80],[89,76,99,67],[92,88,81,75],[78,83,88,91]])
subject_list = ["Kor","Math","Eng","Com"]
subject_list_kr = ["국어","수학","영어","컴"]
student_list = ["Lee","Park","Jung","Choid"]

print(f"총점 반평균 : {score.sum()/4:.2f}")

idx = 0
print("\n각 과목의 평균과 최고점")
for subject in subject_list_kr:
    print(f"{subject}: {score.mean(axis=1)[idx]:.2f},{score.max(axis=1)[idx]}")
    idx +=1

print("\n각 학생별 평균")
idx = 0
for name in student_list:
    print(f"{name} : {score.mean(axis=0)[idx]:.2f}")
    idx +=1
```

```python
plt.title("Average score of Students")
plt.xlabel("Students")
plt.ylabel("Average score")
plt.plot(student_list,score.mean(axis=0),'bp:',lw=1,mec='b',mfc='r',ms=5)
plt.show()
```

```python
plt.title("Average and Top score of Subjects")
plt.xlabel("Subjects")
plt.ylabel("Score")
plt.plot(subject_list,score.mean(axis=1),label = "subject avg")
plt.plot(subject_list,score.max(axis=1),label = "subject max")
plt.legend()
plt.show()
```
