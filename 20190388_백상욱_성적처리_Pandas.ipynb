```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StuData.csv",index_col=0)
df["Com"] = [78,99,90,45,80,75]
#df["평균"] = df.iloc[:,1:].mean(axis=1)
df["평균"] = df.mean(axis=1,numeric_only= True)
df.loc["평균",0] = None
df.loc["평균",1:] = df.mean(axis=0,numeric_only=True)
df.drop(columns=0,inplace = True)
df.round(2)
```

```python
plt.title("과목 점수 평균")
plt.plot(df.mean(axis=0,numeric_only=True).iloc[:-1],'bp:',lw=1,mec='b',mfc='r',ms=5,label="과목 점수 평균")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title("학생 과목 및 평균 점수")
for i in (df.columns[1:]):
    plt.plot(df["이름"].iloc[:-1],df[i].iloc[:-1],'p:',lw=2,mfc='r',label=f"{i}")
plt.legend()
plt.show()
```
