```python
import numpy as np
x = np.linspace(-1.6,1.6,10000)
f = lambda x: np.sqrt(np.cos(x))*np.cos(200*x) + np.sqrt(abs(x))-0.7)*pow((4-x*x),0.01)
plt.plot(x,list(map(f,x)))
plt.show()
```

```python
import numpy as np
X=np.linspace(-np.pi,np.pi,256)
C=np.cos(X)
S=np.sin(X)
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])
plt.yticks([-1,0,1])
plt.show()
```

```python
X = np.linspace(-np.pi,np.pi,256)
S = np.sin(X)
C = np.cos(X)
fig1 = plt.figure()
plt.plot(0,0,c='red',lw=5,marker='o',ms=15)
plt.plot(X,S,label='sine')
plt.legend(loc=0)
fig2 = plt.figure()
plt.plot(X,C,label='cosine')
plt.legend(loc=0)
plt.show()
```

```python
with plt.xkcd():
    plt.title("XKCD style plot!!")
    X = np.linspace(-np.pi,np.pi,256)
    C = np.cos(X)
    plt.plot(X,C)
    plt.scatter([np.pi*2/3],[np.cos(np.pi*2/3)],color='r',lw=10)
    plt.annotate('-0.5 Here!',xy=(np.pi*2/3,np.cos(np.pi*2/3)),xycoords='data',xytext=(0.5,0.05),textcoords="axes fraction",fontsize=16,arrowprops=dict(arrowstyle="->",linewidth=1))
```
