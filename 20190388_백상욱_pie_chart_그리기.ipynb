```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.Series([262,269,162,306],["Java","Python","C++","Ruby"],)

plt.figure(figsize=(7,7))
plt.pie(df,labels=df.index,startangle=45,
       counterclock=True, 
       autopct=lambda p : '{:.1f}%'.format(p),
        explode = [0,0.1,0,0],
        shadow=True,
        colors = ['lightskyblue', 'gold' ,'yellowgreen', 'lightcoral'],
        textprops={'fontsize': 14}
       )
plt.show()
```
