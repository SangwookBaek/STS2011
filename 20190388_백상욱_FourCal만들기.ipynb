```python
class FourCal():
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  def sum(self):
      return self.num1 + self.num2
  def sub(self):
      return self.num1 - self.num2
  def mul(self):
      return self.num1*self.num2
  def div(self):
      return self.num1 / self.num2
```

```python
a = FourCal(4,2)
b = FourCal(3,7)
```

```python
a.sum()
```

```python
a.sub()
```

```python
a.mul()
```

```python
a.div()
```
