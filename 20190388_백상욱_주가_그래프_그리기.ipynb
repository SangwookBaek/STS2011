```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(columns=["open","high","low","close"])
df.loc["16.02.29",:] = [11650,12100,11600,11900]
df.loc["16.02.26",:] = [11100,11800,11050,11600]
df.loc["16.02.25",:] = [11200,11200,10900,11000]
df.loc["16.02.24",:] = [11100,11100,10950,11100]
df.loc["16.02.23",:] = [11000,11150,10900,11050]
```

```python
plt.title("DaeShin")
plt.plot(df.index,df['open'],label = "open")
plt.plot(df.index,df['high'],label = "high")
plt.plot(df.index,df['low'],label = "low")
plt.plot(df.index,df['close'],label = "close")
plt.xlim(xmin=df.index[0],xmax=df.index[-1])
plt.legend()
plt.show()
```

```python
w = 0.15
index = np.arange(5)

plt.title("Daeshin")
# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.barh(index, df['open'],w,label = "open")
b2 = plt.barh(index+ 1*w, df['high'],w,label = "high")

b3 = plt.barh(index+2*w, df['low'],w,label = "low")

b4 = plt.barh(index + 3*w, df['close'],w,label = "close")
# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.yticks(index+1.5*w, df.index)
#plt.xticks(index,df.index)
plt.legend()
plt.show()
```

```python
plt.title("Variation of Closing Price")
plt.plot(df.index,df['close'],label = "close")
plt.xlim(xmin=df.index[0],xmax=df.index[-1])
plt.legend()
plt.show()
```
